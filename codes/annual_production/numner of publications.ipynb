```python
import pandas as pd
import matplotlib.pyplot as plt

# Chart options
font_size = 20
figure_size = (10, 6)
google_colors = {
    'blue': '#4285F4',
    'red': '#DB4437',
    'yellow': '#F4B400',
    'green': '#0F9D58',
}
bar_color = google_colors['green']
figure_background = 'white'

file_path = r'c:\Users\kamalesh\OneDrive\Articles\Review Article\4. ExplainableAI\codes\annual_production\Annual_Production_bibliometrix_2026-05-03.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')

df = df[['Year', 'Articles']].dropna()
df['Year'] = df['Year'].astype(int)
df['Articles'] = df['Articles'].astype(int)
yearly_publications = df.groupby('Year', as_index=False)['Articles'].sum()

plt.rcParams.update({'font.size': font_size})
fig, ax = plt.subplots(figsize=figure_size, facecolor=figure_background)
bars = ax.bar(yearly_publications['Year'].astype(str), yearly_publications['Articles'], color=bar_color)
ax.bar_label(bars, padding=3, fontsize=font_size)
ax.set_xlabel('Year', fontsize=font_size + 1)
ax.set_ylabel('Number of publications', fontsize=font_size + 1)
ax.tick_params(axis='both', labelsize=font_size)
ax.grid(axis='y', linestyle='--', alpha=0.2)
ax.set_axisbelow(True)
ax.set_ylim(0, 40)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
